## Question-1:

Given preorder of a binary tree, calculate its **[depth(or height)](https://www.geeksforgeeks.org/write-a-c-program-to-find-the-maximum-depth-or-height-of-a-tree/)** [starting from depth 0]. The preorder is given as a string with two possible characters.

1. ‘l’ denotes the leaf
2. ‘n’ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can ‘n’ or ‘l’ or mix of both.

**Examples :**

Input  : nlnll
Output : 2
Explanation :

!https://media.geeksforgeeks.org/wp-content/uploads/btree1.png

Input  : nlnnlll
Output : 3

</aside>

```python
max_depth = 0

def calculateDepth(preorder, index, depth):
    global max_depth
    
    if index >= len(preorder) or preorder[index] == 'l':
        return depth
    
    if preorder[index] == 'n':
        depth += 1
        left_depth = calculateDepth(preorder, index + 1, depth)
        right_depth = calculateDepth(preorder, index + 1, depth)
        max_subtree_depth = max(left_depth, right_depth)
        max_depth = max(max_depth, max_subtree_depth)
        return max_subtree_depth

def calculateTreeDepth(preorder):
    global max_depth
    max_depth = 0
    depth = calculateDepth(preorder, 0, 0)
    return max_depth

# Example usage:
preorder = "nlnll"
result = calculateTreeDepth(preorder)
print(result) 

preorder = "nlnnlll"
result = calculateTreeDepth(preorder)
print(result)
```

## Question-2:

Given a Binary tree, the task is to print the **left view** of the Binary Tree. The left view of a Binary Tree is a set of leftmost nodes for every level.

**Examples:**

***Input:***

            4

          /   \

        5     2

             /   \

            3     1

           /  \

          6    7

***Output:** 4 5 3 6*

**Explanation:**

!https://media.geeksforgeeks.org/wp-content/cdn-uploads/left-view.png

***Input:***

                    1

                  /   \

                2       3

                 \

                   4

                     \

                        5

                           \

                             6

**Output:** 1 2 4 5 6

</aside>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def left_view(root):
    if root is None:
        return
    
    queue = []
    queue.append(root)
    left_view_nodes = []

    while queue:
        count = len(queue)
        for i in range(count):
            node = queue.pop(0)

            if i == 0:
                left_view_nodes.append(node.data)

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
    
    for node in left_view_nodes:
        print(node, end=" ")
```

```python
root = Node(4)
root.left = Node(5)
root.right = Node(2)
root.right.left = Node(3)
root.right.right = Node(1)
root.right.left.left = Node(6)
root.right.left.right = Node(7)
left_view(root)
```

## Question-3:

Given a Binary Tree, print the Right view of it.

The right view of a Binary Tree is a set of nodes visible when the tree is visited from the Right side.

**Examples:**

**Input:**

         1

      /     \

   2         3

/   \       /  \

4     5   6    7

             \

               8

**Output**: 

Right view of the tree is 1 3 7 8

**Input:**

         1

       /

    8

  /

7

**Output**: 

Right view of the tree is 1 8 7

</aside>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def right_view(root):
    if root is None:
        return
    
    queue = []
    queue.append(root)
    right_view_nodes = []

    while queue:
        count = len(queue)
        for i in range(count):
            node = queue.pop(0)

            if i == count - 1:
                right_view_nodes.append(node.data)

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
    
    for node in right_view_nodes:
        print(node, end=" ")
```

```python
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)
root.right.right.right = Node(8)
right_view(root)
```

## Question-4:

Given a Binary Tree, The task is to print the **bottom view** from left to right. A node **x** is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

**Examples:**

**Input:**

             20

           /     \

        8         22

    /      \         \

5         3        25

        /    \

   10       14

**Output:** 5, 10, 3, 14, 25.

**Input:**

             20

           /     \

        8         22

    /      \      /   \

 5         3    4     25

         /    \

     10       14

**Output:**

5 10 4 14 25.

**Explanation:**

If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the level traversal.

**3 and 4** are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.

</aside>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def bottom_view(root):
    if root is None:
        return
    
    queue = []
    queue.append((root, 0))  
    bottom_view_map = {}

    while queue:
        node, hd = queue.pop(0)

        bottom_view_map[hd] = node.data

        if node.left:
            queue.append((node.left, hd - 1))
        if node.right:
            queue.append((node.right, hd + 1))
    
    for hd in sorted(bottom_view_map):
        print(bottom_view_map[hd], end=" ")
```

```python
root = Node(20)
root.left = Node(8)
root.right = Node(22)
root.left.left = Node(5)
root.left.right = Node(3)
root.right.right = Node(25)
root.left.right.left = Node(10)
root.left.right.right = Node(14)
bottom_view(root)
```
